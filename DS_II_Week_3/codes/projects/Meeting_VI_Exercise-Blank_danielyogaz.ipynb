### _Cluster_
#### _Load_ data `concerete_data.csv`

```python
import pandas as pd

data = pd.read_csv('..//..//data//input//concerete_data.csv')
data.head()
```

#### Lakukan _scaling_ pada data dengan _Min-Max scaler_

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(data)
data_scaled = pd.DataFrame(scaler.transform(data), columns=data.columns)
data_scaled.head()
```

#### Lakukan _clustering_ dengan metode hirarki pada data yang sudah di-_scaling_. Simpan hasil _clustering_ pada kolom baru pada `data` yang dinamai sebagai `cluster1`

```python
from sklearn.cluster import AgglomerativeClustering

agglo = AgglomerativeClustering(n_clusters=4)
data['cluster1'] = agglo.fit(data_scaled)
data.head()
```

#### Lakukan _clustering_ dengan metode partisi pada data yang sudah di-_scaling_. Simpan hasil _clustering_ pada kolom baru pada `data` yang dinamai sebagai `cluster2`

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(data_scaled)
data['cluster2'] = kmeans.fit(data_scaled)
data.head()
```

#### Lihat hasil _cluster1_ dan _cluster2_. Tentukan _cluster1_ mana yang bersesuaian dengan _cluster2_ (semisal _cluster_ 1 pada _cluster1_ sama dengan _cluster_ 3 pada _cluster2_). Gunakan `pd.crosstab`

```python
pd.crosstab(data['cluster1'], data['cluster2'])
```

### _Association rule_
#### _Load_ data pada `items.xlsx`

```python
data= pd.read_excel('..//..//data//input//items.xlsx')
data.head()
```

#### Data sudah dalam bentuk _one hot encoding_. Dapatkan _frequent itemset_ dengan _support_ minimum sebesar 10%

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(data, min_support=0.10, use_colnames=True)
frequent_itemsets
```

#### Dari _frequent itemsets_ di atas, dapatkan _rules_ dengan:
1. _Confidence_ > 0.8
2. _Leverage_ > 0.075
3. _Conviction_ > 2.5

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules1
```

```python
rules2 = association_rules(frequent_itemsets, metric="leverage", min_threshold=0.075)
rules2
```

```python
rules3 = association_rules(frequent_itemsets, metric="conviction", min_threshold=2.5)
rules3
```
